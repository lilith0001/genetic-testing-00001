```python
import json
import numpy as np
import random
import copy
from math import trunc

# dictionary setup
sharpness_raw      = {'red':0.50, 'orange':0.75 , 'yellow':1.00, 'green':1.05, 'blue':1.20,   'white':1.32}
sharpness_element  = {'red':0.25, 'orange':0.50,  'yellow':0.75, 'green':1.00, 'blue':1.0625, 'white':1.15}
skills = {'Resuscitate'     :{1:5,    2:10,   3:20}, # critical draw, punishing draw
          'Peak Performance':{1:5,    2:10,   3:20},
          'Critical Boost'  :{1:0.30, 2:0.35, 3:0.40},
          'Weakness Exploit':{1:0.15, 2:0.30, 3:0.50},
          'Resentment'      :{1:5,    2:10,   3:15,   4:20,   5:25},
          'Dragonheart'     :{1:0,    2:0,    3:0,    4:0.05, 5:0.10},
          'Heroics'         :{1:0,    2:0.05, 3:0.05, 4:0.10, 5:0.30},
          'Critical Eye'    :{1:0.05, 2:0.10, 3:0.15, 4:0.20, 5:0.25, 6:0.30, 7:0.40},
          'Attack Boost'    :{1:{'+':3, '*':0},    2:{'+':6, '*':0},    3:{'+':9,  '*':0},    4:{'+':7,  '*':0.05}, 5:{'+':8,  '*':0.06}, 6:{'+':9, '*':0.08}, 7:{'+':10, '*':0.10}},
          'Agitator'        :{1:{'+':4, '*':0.03}, 2:{'+':8, '*':0.05}, 3:{'+':12, '*':0.07}, 4:{'+':16, '*':0.10}, 5:{'+':20, '*':0.15}}}
decorations = {'Flawless Jewel 2'  :{'skill':'Peak Performance','level':2},
               'Tenderizer Jewel 2':{'skill':'Weakness Exploit','level':2},
               'Critical Jewel 2'  :{'skill':'Critical Boost',  'level':2},
               'Expert Jewel 2'    :{'skill':'Critical Eye',    'level':2},
               'Attack Jewel 2'    :{'skill':'Attack Boost',    'level':2},
               'Crisis Jewel 2'    :{'skill':'Resuscitate',     'level':2},
               'Furor Jewel 2'     :{'skill':'Resentment',      'level':2},
               'Challenger Jewel 2':{'skill':'Agitator',        'level':2},
               'Potential Jewel 2' :{'skill':'Heroics',         'level':2}}

with open('sets.json') as f:
    set_dictionary = json.load(f)
set_names = []
for key, value in set_dictionary.items():
    set_names.append(key)
deco_keys = list(decorations.keys())
    
def print_set(set):
    print()
    print('!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!')
    print()
    print(f'reached epoch {epoch} after {result[1]} generations.')
    print()
    print("['efr'] -> ", trunc(set['efr']))
    print("['skills']")
    for skill in set['skills']:
        if skill in skills:
            print('\t',skill, set['skills'][skill])
    for skill in set['skills']:
        if skill not in skills:
            print('\t',skill, set['skills'][skill])
    print("['equipment']")
    for piece in set['equipment']:
        print(f"\t['{piece}']")
        if piece == 'weapon':
            print(f"\t\tattack      -> {set['equipment'][piece]['attack']}")
            print(f"\t\taffinity    ->  {trunc(set['equipment'][piece]['affinity'] * 100)}%")
            print(f"\t\tsharpness   -> {set['equipment'][piece]['sharpness']}")
            print(f"\t\tslots       -> {set['equipment'][piece]['slots']}")
        else:
            if piece != 'talisman':
                print(f"\t\tname   -> {set['equipment'][piece]['name']}")
            # printing equipment skills
            print("\t\tskills -> ", end='')
            try:
                i = 0
                for skill in set['equipment'][piece]['skills']:
                    if i < len(set['equipment'][piece]['skills']) - 1:
                        print(skill, set['equipment'][piece]['skills'][skill], end=', ')
                        i = i + 1
                    else:
                        print(skill, set['equipment'][piece]['skills'][skill])
            except KeyError:
                print('None')
            print("\t\tslots  ->",set['equipment'][piece]['slots'])
        # printing decorations
        print("\t\tdecorations -> ", end='')
        try:
            for i in range(0, len(set['equipment'][piece]['decorations'])):
                if i < len(set['equipment'][piece]['decorations']) - 1:
                    print(set['equipment'][piece]['decorations'][i], end=', ')
                else:
                    print(set['equipment'][piece]['decorations'][i])
        except KeyError:
            print('None')
        
def sort_generation(gen):
    # bubble sort
    for set in gen['sets']:
        for j in range(len(gen['sets']) - 1):
            if gen['sets'][j]['efr'] < gen['sets'][j+1]['efr']:
                gen['sets'][j], gen['sets'][j+1] = gen['sets'][j+1], gen['sets'][j]
    None

def populate_skill_info(set):
    set['skills'] = dict()
    # get all skills from equipment
    for piece in set['equipment']:
        if piece == 'weapon':
            None
        else:
            for skill in set['equipment'][piece]['skills']:
                try:
                    # if the current skill level is less than the maximum skill level
                    if  set['skills'][skill] < len(skills[skill]):
                        set['skills'][skill] += 1
                except KeyError:
                    set['skills'][skill] = 1
        # get all skills from decorations
        try:
            for deco in set['equipment'][piece]['decorations']:
                try:
                    # if the current skill level is less than the maximum skill level
                    if set['skills'][decorations[deco]['skill']] < len(skills[decorations[deco]['skill']]):
                        set['skills'][decorations[deco]['skill']] += 1
                except KeyError:
                    set['skills'][decorations[deco]['skill']] = 1
        except KeyError:
            None
    None
    
def generate_random_decorations_piece(set, piece):
    set['equipment'][piece]['decorations'] = list()
    for slot in set['equipment'][piece]['slots']:
        if slot > 0:
            random.shuffle(deco_keys)
            for deco in deco_keys:
                if decorations[deco]['level'] <= slot:
                    set['equipment'][piece]['decorations'].append(deco)
                    break
                            
def generate_random_set_decorations(set):
    for piece in set['equipment']:
        generate_random_decorations_piece(set, piece)
           
def generate_random_set_piece(set, piece):
    while True:
        set['equipment'][piece] = copy.deepcopy(set_dictionary[set_names[np.random.randint(0,145)]][piece])
        if set['equipment'][piece] != None: break

def generate_random_set_pieces(set):
    for piece in set['equipment']:
        if piece == 'weapon' or piece == 'talisman': None
        else: generate_random_set_piece(set, piece)

def initialize_set():
    set = {'skills':dict(),
           'equipment':{'weapon'  :copy.deepcopy(weapon),
                        'talisman':copy.deepcopy(talisman),
                        'helm' :dict(),
                        'chest':dict(),
                        'arms' :dict(),
                        'waist':dict(),
                        'legs' :dict()}}
    return set

def generate_random_set(weapon, talisman):
    set = initialize_set()
    generate_random_set_pieces(set)
    generate_random_set_decorations(set)
    populate_skill_info(set)
    return set
    
# fitness function
def calc_efr(set):
    base_raw = set['equipment']['weapon']['attack']
    bonus_raw = 0
    sharpness = sharpness_raw[set['equipment']['weapon']['sharpness']]
    affinity  = set['equipment']['weapon']['affinity']
    critical_multiplier = 0.25
    for skill in set['skills']:
        if   skill == 'Resuscitate' or skill == 'Resentment':
             bonus_raw = bonus_raw + skills[skill][set['skills'][skill]]
        elif skill == 'Peak Performance':
             if 'Dragonheart' not in set['skills']:
                bonus_raw = bonus_raw + skills[skill][set['skills'][skill]]
        elif skill == 'Dragonheart':# or skill == 'Heroics' :
             bonus_raw = bonus_raw + base_raw * skills[skill][set['skills'][skill]]
        elif skill == 'Critical Eye' or skill == 'Weakness Exploit':
             affinity = affinity + skills[skill][set['skills'][skill]]
        elif skill == 'Critical Boost':
             critical_multiplier = skills[skill][set['skills'][skill]]
        elif skill == 'Attack Boost':# or skill == 'Agitator':
             bonus_raw = bonus_raw + skills[skill][set['skills'][skill]]['+']
             bonus_raw = bonus_raw + base_raw * skills[skill][set['skills'][skill]]['*']
    critical_modifier = 1 + min(affinity, 1) * critical_multiplier
    efr = (base_raw + bonus_raw) * sharpness * critical_modifier
    set['efr'] = efr
        
# selection function - sort the sets by EFR
def select(gen, verbose):
    sum = 0
    for set in gen['sets']:
        calc_efr(gen['sets'][set])
        sum += gen['sets'][set]['efr']
    gen['avg'] = sum / len(gen['sets'])
    sort_generation(gen)

def generate_child(parent0, parent1):
    child = initialize_set()
    for piece in child['equipment']:
        if(random.random() > 0.5):
              child['equipment'][piece] = copy.deepcopy(parent0['equipment'][piece])
        else: child['equipment'][piece] = copy.deepcopy(parent1['equipment'][piece])
    return(child)
    
def crossover(weapon, talisman, gen, gen_size, verbose):   
    # uniform crossover,  each chromosome has equal chance
    # the current best set will produce more sets than others.
    # the efr of the best set / the average set will determine
    # the percentage of the next generation from the best.
    ratio = gen['sets'][0]['efr'] / gen['avg']
    child_count = trunc(gen_size * trunc((ratio - 1) * 100) / 100) * 0.5
    if verbose == 1:
        print('  [crossover]')
        print(f'    the current best set will have {child_count} children')
    next_gen = {'sets':dict()}
    valstrax_count = (gen_size - child_count) / 2
    for i in range(0, gen_size):
        # 50% of best children from second best
        if i < trunc(child_count / 2):
            next_gen['sets'][i] = generate_child(gen['sets'][0], gen['sets'][1])
        # 25% of best children from third best
        elif i < trunc(child_count / 2) + trunc(child_count / 4):
            next_gen['sets'][i] = generate_child(gen['sets'][0], gen['sets'][2])
        # 25% of best children from fourth best
        elif i < child_count:
            next_gen['sets'][i] = generate_child(gen['sets'][0], gen['sets'][3])
        elif i < valstrax_count + child_count:
            next_gen['sets'][i] = valstrax_set(weapon, talisman)
        else:
            next_gen['sets'][i] = generate_random_set(weapon, talisman)
    return next_gen, child_count

def mutate_set(set, x):
    for piece in set['equipment']:
        if piece == 'weapon' or piece == 'talisman':
             None
        else:
            if   x <= 1 / 5:
                #gen['sets'][set]['equipment'][piece] = dict()
                generate_random_set_piece(set, piece)
                break
            elif x > 1 / 5 and x <= 2 / 5:
                #gen['sets'][set]['equipment'][piece]
                generate_random_set_piece(set, piece)
                break
            elif x > 2 / 5 and x <= 3 / 5:
                #gen['sets'][set]['equipment'][piece]
                generate_random_set_piece(set, piece)
                break
            elif x > 3 / 5 and x <= 4 / 5:
                #gen['sets'][set]['equipment'][piece]
                generate_random_set_piece(set, piece)
                break
            elif x > 4 / 5:
                #gen['sets'][set]['equipment'][piece]
                generate_random_set_piece(set, piece)
                break
    # 50% chance of also changing the decorations on a piece
    y = random.random()
    z = random.random()
    if y < 0.5:
        if   z <= 1 / 7:
            generate_random_decorations_piece(set, 'weapon')
        elif z > 1 / 7 and z <= 2 / 7:
            generate_random_decorations_piece(set, 'talisman')
        elif z > 2 / 7 and z <= 3 / 7:
            generate_random_decorations_piece(set, 'helm')
        elif z > 3 / 7 and z <= 4 / 7:
            generate_random_decorations_piece(set, 'chest')
        elif z > 4 / 7 and z <= 5 / 7:
            generate_random_decorations_piece(set, 'arms')
        elif z > 5 / 7 and z <= 6 / 7:
            generate_random_decorations_piece(set, 'waist')
        elif z > 6 / 7:
            generate_random_decorations_piece(set, 'legs')
                
def mutate_gen(gen, verbose):
    # mutation function. one armor piece or one decoration 
    # for each set in the generation will mutate
    for set in gen['sets']:
        mutate_set(gen['sets'][set], random.random())
            
def gen_gen(weapon, talisman, gen, gen_size, improvement_delta, verbose):
    current_optimum = copy.deepcopy(gen['sets'][0])
    print('current optimum: ', current_optimum['efr'])
    gen, child_count = crossover(weapon, talisman, gen, gen_size, verbose)
    mutate_gen(gen, verbose)
    for set in gen['sets']:
        populate_skill_info(gen['sets'][set])
    select(gen, verbose)
    if gen['sets'][0]['efr'] <= current_optimum['efr']:
        gen['sets'][0] = copy.deepcopy(current_optimum)
            
        if verbose == 1:
            print('239 top 3 efr:')
            for i in range(0, 3):
                print('     ',trunc(gen['sets'][i]['efr']))
            
        improvement_delta = improvement_delta + 1
        if verbose == 1:
            print('  no improvement..')
            print()
    else:
        improvement_delta = 0
        if verbose == 1:
            print('  improvement!')
            print()
    return gen, improvement_delta

def random_initial_set(weapon, talisman, gen_size):
    gen = {'sets':dict()}
    for i in range(0,gen_size):
        gen['sets'][i] = generate_random_set(weapon, talisman)
    select(gen, 0)
    return gen

def valstrax_set(weapon, talisman):
    valstrax = dict()
    valstrax['equipment'] = copy.deepcopy(set_dictionary['Valstrax'])
    valstrax['equipment']['weapon']    = copy.deepcopy(weapon)
    valstrax['equipment']['talisman']  = copy.deepcopy(talisman)
    generate_random_set_decorations(valstrax)
    populate_skill_info(valstrax)
    return valstrax
```

```python
def evolve(weapon, talisman, epoch, gen_size, verbose):
    gen = random_initial_set(weapon, talisman, gen_size)
    improvement_delta = 0
    generation_count = 0
    while True:
        if verbose == 1:
            print(f'[generation {generation_count}]')
        gen, improvement_delta = gen_gen(weapon, talisman, gen, gen_size, improvement_delta, verbose)
        generation_count = generation_count + 1
        if improvement_delta >= epoch:
            return gen['sets'][0], generation_count
            break
```

```python
weapon = {'attack'     :190,
          'affinity'   :0.40,
          'sharpness'  :'white',
          'slots'      :[2,0,0],
          'decorations':list()}
talisman = {'skills':{'Speed Eating':1, 'Weakness Exploit':1},
            'slots' :[2, 2, 0]}

# epoch, gen_size, print (0 = no print, 1 = print)
epoch    = 5
gen_size = 1000
verbose  = 0
result = evolve(weapon, talisman, epoch, gen_size, verbose)
print_set(result[0])
```
